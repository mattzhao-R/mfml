# Lecture 7, the SVD

```python
import numpy as np
import scipy.linalg as sla
import numpy.linalg as la
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

## create and plot data

```python
n = 3
p = 100
S0 = np.diag((np.array([10, 2, .2])))/10
U0 = sla.orth(np.random.rand(n,n))
V0 = np.random.randn(p,n)

X = U0@S0@V0.T
print(la.matrix_rank(X))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
p = ax.scatter(X[0], X[1], X[2])
plt.show()
```

## compute SVD and plot corresponding basis vectors

```python
U, S, V = la.svd(X,full_matrices=False)

print(U.shape)

fig = plt.figure()
ax = plt.axes(projection='3d')

p = ax.scatter(X[0], X[1], X[2])
ax.plot3D((0,U[0,0]),(0,U[1,0]),(0,U[2,0]),'green',linewidth=3)
ax.plot3D((0,U[0,1]),(0,U[1,1]),(0,U[2,1]),'blue',linewidth=3)
ax.plot3D((0,U[0,2]),(0,U[1,2]),(0,U[2,2]),'red',linewidth=3)


fig = plt.figure()
ax = plt.axes(projection='3d')

print(S)

p = ax.scatter(X[0], X[1], X[2])
ax.plot3D((0,U[0,0]*S[0]),(0,U[1,0]*S[0]),(0,U[2,0]*S[0]),'green',linewidth=3)
ax.plot3D((0,U[0,1]*S[1]),(0,U[1,1]*S[1]),(0,U[2,1]*S[1]),'blue',linewidth=3)
ax.plot3D((0,U[0,2]*S[2]),(0,U[1,2]*S[2]),(0,U[2,2]*S[2]),'red',linewidth=3)
plt.show()

```
