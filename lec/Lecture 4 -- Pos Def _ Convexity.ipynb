# Positive definiteness and convexity

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

## Construct function $$f(w) = w^\top Q w$$ and plot it

```python
n = 101 # number of grid points in each direction for plotting function 

# Q = np.array([[1, 0], [0, 1]])
# Q = np.array([[-1, 0], [0, -1]])
Q = np.array([[1, 0], [0, -1]])

w1, w2 = np.meshgrid(np.linspace(-1.5, 1.5, n), np.linspace(-1.5, 1.5, n))
f = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        wij = np.array([w1[i, j], w2[i, j]])
        f[i, j] = np.dot(wij, Q).dot(wij)

ax = plt.gca(projection='3d')
ax.plot_surface(w1, w2, f, cmap=cm.viridis)
plt.title('plot of f(w) = w^T Q w')
ax.set_xlabel('w_1')
ax.set_ylabel('w_2')
ax.set_zlabel('f(w)')



plt.show()
```
