```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

### Subspace projection in 3d

```python
# activate interactive plotting:
%matplotlib notebook
# note it's not supported in colab, use %matplotlib inline  instead (but it's not interactive)

fig = plt.figure(1)
ax = fig.add_subplot(projection='3d')

U = np.random.rand(3, 2)
# U = np.array([[1,1],[-1,1],[0,0]])
# U = np.array([[1,1],[-1,1],[1,-1]])
V = np.random.randn(2, 100)

X = U @ V
ax.scatter(X[0, :], X[1, :], X[2,: ]) # rows of matrix X

y = 6 * (np.random.rand(3) - 0.5)
# vy1 = np.random.randn(2)
# y = U@vy1 + np.random.randn(3)
ax.plot((y[0], 0),(y[1], 0),(y[2], 0),'r', linewidth=3, label='y')

vy = la.inv(U.T@U)@U.T@y
projection = U@vy # find projection

# make plots
ax.plot((projection[0], 0),(projection[1], 0),(projection[2], 0),'g',linewidth=3,
        label='projection of y onto subspace')
ax.plot((projection[0], y[0]),(projection[1], y[1]),(projection[2], y[2]),'b--',linewidth=3,
        label='projection residual')
plt.title("Subspace projection illustration ")

plt.legend();
```
